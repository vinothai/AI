```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4, 5], [2, 4, 6, 8, 16], 'b*-', label='Major')
plt.plot([1, 3, 5, 7, 9], [0, 6, 10, 14, 18], 'ro-', label='Minor')
plt.title('pyplot')
plt.xlabel('Some x (unit)')
plt.ylabel('Some y (unit)')
plt.axis([0, 20, 0, 20])
plt.legend()

plt.show()
```
