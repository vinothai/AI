```python
import matplotlib.pyplot as plt

plt.figure(1)
plt.subplot(2, 1, 1)
plt.plot([1, 2, 3, 4, 5], [2, 4, 6, 8, 16], 'b*-', label='Major')
plt.title('Sub plot 1')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot([1, 3, 5, 7, 9], [0, 6, 10, 14, 18], 'ro--', label='Minor')
plt.title('Sub plot 2')
plt.legend()

plt.tight_layout()

plt.show()
```
